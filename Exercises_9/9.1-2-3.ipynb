```python
import random
import matplotlib.pyplot as plt
import numpy as np
import math

def rand_range (xMin, xMax) :
    '''
    generazione di un numero pseudo-casuale distribuito fra xMin ed xMax
    '''
    return xMin + random.random () * (xMax - xMin)

tau = 5 #characteristic time
list=[]
def func(x): return -((np.log(1-x))*(tau))

for i in range(100000):
    list.append(func(rand_range(0,1)))

x_coord=np.linspace(0, 60, 100000)
y_coord=((1/tau)*np.exp(-(x_coord)/tau))

fig, ax = plt.subplots (nrows = 1, ncols = 1)
ax.hist (list,
         color = 'orange',
         bins=int(np.ceil(1+3.322*np.log(100000))),
         density=True
        )
ax.plot(x_coord, y_coord, color = 'blue')
plt.show()
```
