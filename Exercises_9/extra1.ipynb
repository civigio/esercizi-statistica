```python
import random
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def rand_range (xMin, xMax) :
    '''
    generazione di un numero pseudo-casuale distribuito fra xMin ed xMax
    '''
    return xMin + random.random () * (xMax - xMin)

def func(x, tau): return -((np.log(1-x))*(tau))
```

```python
def likelihood(list):
    return np.sum(list)
```

```python
likelihood_values = []
list = []
x_coord=np.arange(0.1, 10, 0.1)
y_coord=[]
tau = 4.3

for _ in range(10000):
    list.append(rand_range(0,1))

exponential_pdf=func(np.array(list), tau)

for value in x_coord:
    exponential=((1/value)*np.exp(-(exponential_pdf)/value))
    y_coord.append(likelihood(np.log(exponential)))

fig, ax = plt.subplots(1, 1)
plt.plot(x_coord, y_coord)
plt.grid()
plt.show()
```
